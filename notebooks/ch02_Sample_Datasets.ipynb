```python
#now adding another real world dataset, but a regression one
#the boston housing dataset used in the book was unavaliable, so i changed to the california one,
#also changing other parts of the code for it to make sense
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import PolynomialFeatures

california = fetch_california_housing()
print("Data shape: ", california.data.shape)
```

```python
poly = PolynomialFeatures(degree=2, include_bias=False)
X = poly.fit_transform(california.data)

y = california.target

print("X.shape:", X.shape)
```

```python
from sklearn.model_selection import train_test_split
X, y = mglearn.datasets.make_forge()

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=3)
```

```python
clf.fit(X_train, y_train)
```

```python
print("Test set predictions:" , clf.predict(X_test))
```

```python
print("Test set accuracy: {:.2f}".format(clf.score(X_test, y_test)))
```
